```python
from openai import OpenAI
```

```python
client = OpenAI()
```

```python
input_text = None

with open("example_input.txt", "r") as file:
    input_text = file.read()
```

```python
print(repr(input_text))
```

```python

agentic_prompt = "".join(["You are a legal professional writing a case summary of the following document. ",
                          "The case summary should start with one sentence that states what the court does, ",
                          "followed by one or two paragraphs describing the facts. ",
                          "Then the summary should describe the judicial history in two paragraphs. ",
                          "Lastly, the summary will state the court's decision in one sentence, ",
                          "followed by an optional paragraph explaining what the public should learn from this case."])

prompt = agentic_prompt

llm_input = f"{prompt}\n\n{repr(input_text)}"

response = client.chat.completions.create(
            model="gpt-4o",
            messages=[
                {
                "role": "user",
                "content": [
                    {
                    "type": "text",
                    "text": llm_input
                    }
                ]
                }
            ],
            temperature=1,
            max_tokens=2048,
            top_p=1,
            frequency_penalty=0,
            presence_penalty=0,
            response_format={
                "type": "text"
            }
        )
output = response.choices[0].message.content
print(output)
```
